##### - Using 3 asterisks or 3 underscores under the text

- Example with three asteriks:

    - I like strawberries

- Example with three underscores:     

    - I like strawberries 
___

- Not possible with 3 dots:

    - I like strawberries     
... 


# 4. Bold and italics

##### - Using x number of asterisks, underscores  or specific words before and after the word or sentence

1. Bold: Using 2 asterisks or 2 underscores or "strong" between symbols "<>" at the beginning and "strong" between "</>" at the end

- With **:
    - **Strawberry in bold**                                   
- With __:
    - __Strawberry in bold__                                       
- With "strong" between <>:
    - <strong>Strawberry in bold</strong>                          

2. Italics: Using 1 asterisk or 1 underscore or "em" between symbols "<>" at the beginning and "em" between "</>" at the end

- With *:
    - *Strawberry in italics*       
- With _:                                
    - _Strawberry in italics_ 
- With "em" between <>:                                   
    - <em>Strawberry in italics</em>                               

3. Bold and Italics: Using 3 asterisks or 3 underscores or "strong" + "em" between symbols "<>" at the beginning and "strong" + "em" between "</>" at the end

- With ***:
    - ***Strawberry in bold and italics1***            
- With ___:
    - ___Strawberry in bold and italics2___     
- With "strong" + "em" between <>:                 
    - <strong><em>Strawberry in bold and italics3</strong></em>      

# 5. Highlight a word in a text

##### - Using grave accent " ` "before and after the word 

- Highlighting word "pigs":

    - Once upon a time there was an old mother pig who had three little `pigs` and not enough food to feed them. So when they were old enough, she sent them out into the world to seek their fortunes.

- Without highlighting: 

    - Once upon a time there was an old mother pig who had three little pigs and not enough food to feed them. So when they were old enough, she sent them out into the world to seek their fortunes.

# 7. Add links

##### - Word or sentence to be clicked must be enclosed in [ ]. Url can be added in different ways
 


1. Url added between "( )" next to the word to be clicked

    - Click [here](https://markdown.es/sintaxis-markdown/) to see content      

2. One or more urls can be placed after text using square brackets + colon + url

    - I like strawberries and apples. I will buy some on the local [greengrocer].  
    
    - Dinosaurs were a very large and varied group of prehistoric [reptiles] that dominated the Earth millions of years ago, long before humans were even about to appear. These animals populated every [habitat] on the planet: the oceans, the dry land and even soared through the air, over 135 million years, from their appearance in the Upper Triassic period until their extinction at the end of the Cretaceous period.
    
[greengrocer]:https://www.rockandfruit.com/
[reptiles]:https://www.savalli.us/BIO113/Labs/09.Reptiles.html
[habitat]:https://www.museeprehistoire.com/en/mille-millenaires/le-prehistosite-pour-vivre-la-prehistoire/cinq-habitats-et-un-dolmen-pour-explorez-la-prehistoire   


3. A link can be displayed directly in the text

    - For more information, please enter here https://www.markdownguide.org/
